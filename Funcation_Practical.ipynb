   - List and tuples are both sequence types in Python, but they differ in one key aspect: mutability. Let me demonstrate this with examples:

   # Create a list (mutable)
     my_list = [1, 2, 3, 4, 5]

     # Create a tuple (immutable)
     my_tuple = (1, 2, 3, 4, 5)

     print("Original list:", my_list)
     print("Original tuple:", my_tuple)

   # Demonstration of list and tuple mutability

# Create our sequences
     fruits_list = ["apple", "banana", "cherry"]
     fruits_tuple = ("apple", "banana", "cherry")

     print("Original list:", fruits_list)
     print("Original tuple:", fruits_tuple)

# Modify the list - this works
     fruits_list[1] = "blueberry"
     print("\nAfter list modification:", fruits_list)

# Try to modify the tuple - this will fail
     try:
     print("\nAttempting to modify tuple...")
     fruits_tuple[1] = "blueberry"
except TypeError as e:
     print("Error occurred:", e)
     print("Tuples are immutable and cannot be changed after creation")

     print("\nFinal list:", fruits_list)
     print("Final tuple:", fruits_tuple)

#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

  - # Get input from user
     num = int(input("Enter a positive integer: "))

     # Initialize variables
     factorial = 1
     i = 1

     # Calculate factorial using while loop
     while i <= num:
     factorial *= i
     i += 1

     print(f"The factorial of {num} is {factorial}")
